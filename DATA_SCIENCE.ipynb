```python
import matplotlib.pyplot as plt
import seaborn as sns

#bar cHart for grade distribution
sns.countplot(x='Grade',data=df)
plt.title('Grade Distribution')
plt.xlabel('Grade')
plt.ylabel('Count')
plt.show()
```

```python
#pie chart for gender distribution
gender_counts=df['Gender'].value_counts()
plt.pie(gender_counts,labels=gender_counts.index,autopct='%1.1f%%')
plt.title('Gender Distribution')
plt.show()
```

```python
#histogram for age
plt.hist(df['Age'],bins=10)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
#for now,let's simulate a 'Score' column
df["Score"] = [85,78,90,70]
plt.figure(figsize=(6,4))
sns.scatterplot(x="Age",y="Score",hue = "Gender" , style="Grade",s=100,data=df)
plt.title('Age vs Score')
plt.xlabel('Age')
plt.ylabel('Score')
plt.show()
```

```python
#data Engineering
```

```python
#Feature:Age_Group1(Age<22 else Senior)
df['Age_Group'] = df['Age'].apply(lambda x: 'Junior' if x < 22 else 'Senior')
df.head()
```

```python
#Score Category
df["Score_Category"]=df["Score"].apply(lambda x: "High" if x>80 else "Low")
df.head()
```

```python
#Is_Top_Category
df["Is_Top_Category"]=df["Grade"].apply(lambda x: 1 if x=="A" else 0)
df.head()
```
